## 1.1 Solving Linear Equations

```python
import numpy as np
```

```python
A = np.array(([4,-5],[-2,3]))
print(A)
```

```python
b = np.array(([-13],[9]))
print(b)
```

```python
#A의 역행렬에 b내적
x = np.linalg.inv(A).dot(b)
print(x)
```

```python
theta = 90/180*np.pi

print(theta)

R = np.matrix([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta), np.cos(theta)]])

x = np.matrix([[1], [0]])

y = R*x

print(y)
```

# 4. Eigenvalue and Eigenvector

행렬과 선형 변환은 같은거다.

선형 변환이 주어졌을때 행렬을 찾는 방법

행렬이 주어졌을때 해당하는 선형 변환을 찾는 방법

## 4.1 How to Compute Eigenvalue & Eigenvector

```python
import numpy as np

A = np.array([[1, 0],
              [0, 0]])
D, V = np.linalg.eig(A)

print('D :', D)
print('V :', V)
```
